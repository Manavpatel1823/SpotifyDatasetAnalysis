```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('D:/spotifyDataset/tracks.csv')
df_tracks.head()
```

```python
#check for null values.
pd.isnull(df_tracks).sum()
```

```python
#find the least popular song.
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
```

```python
#describe give us math values and transpose is a matrix type.
df_tracks.describe().transpose()
```

```python
#we use query for condition.
df_tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending = False).head(10)
```

```python
df_tracks.set_index('release_date', inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head(10)
```

```python
df_tracks[["artists"]].iloc[10]
```

```python

df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x : round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head(10)
```

```python
corr_df = df_tracks.drop(["key","mode","explicit"], axis = 1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True,fmt = ".1g",vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths=1, linecolor="Black")
heatmap.set_title("correlation between heatmap variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

```python
#plot for data plot.
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x = "energy", y = "loudness", color = 'c').set(title = "Loudness vs Energy")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, x = "acousticness", y = "popularity", color = 'b').set(title = "Popularity vs Acousticness")
```

```python
df_tracks['dates'] = df_tracks.index.get_level_values('release_date')
df_tracks.dates = pd.to_datetime(df_tracks.dates)
years = df_tracks.dates.dt.year
```

```python
sns.displot(years, discrete=True, aspect = 2, height = 5, kind = "hist").set(title="Number of songs per year")
```

```python
#new csv file read.
df_genre = pd.read_csv("D:/spotifyDataset/SpotifyFeatures.csv")
df_genre.head(5)
```

```python
plt.title("Duration of the songs.")
sns.color_palette("rocket", as_cmap = True)
sns.barplot(y = "genre", x = "duration_ms", data = df_genre)
plt.xlabel("Duration in milisecond.")
plt.ylabel("Genres")
```

```python
plt.figure(figsize=(10,6))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(x = "popularity", y = "genre", data = famous).set(title = "Top 5 genres by popularity.")
```
